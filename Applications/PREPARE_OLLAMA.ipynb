# Ollama 后端准备指南 (Notebook)

用于在应用类 notebook 中使用 Ollama 作为 LLM 后端。


## 1. 安装 Ollama
- 下载并安装: https://ollama.ai/download
- 安装完成后验证:
  - `ollama --version`


![Ollama version check](assets/version.png)


## 2. 拉取模型（统一使用）
- `ollama pull qwen3:4b`


## 3. 启动服务
通常安装后会自动启动服务；若没有，请手动运行:
- `ollama serve`


## 4. 检查服务是否可用
任选其一:
- `ollama list`
- 访问 `http://localhost:11434/api/tags`

![Ollama status](assets/status.png)



## 5. 在 notebook 中使用
示例（与 Apps 里的默认写法一致）:
```python
llm = get_llm_backend("ollama", model="qwen3:4b")
```
